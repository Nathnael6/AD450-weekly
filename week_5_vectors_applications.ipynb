2. Non-mean centered vectors

Run all three tests on the two test cases and describe what you find.

````python
import numpy as np
import numpy.typing as npt
from scipy import spatial


def pearson_correlation(
    np_vector_a: npt.NDArray,
    np_vector_b: npt.NDArray
) -> np.float64:
    # Compute means
    mean_a = np.mean(np_vector_a)
    mean_b = np.mean(np_vector_b)

    # Compute the numerator
    numerator = np.sum(
        (np_vector_a - mean_a) * (np_vector_b - mean_b)
    )

    # Compute the denominator
    denominator = np.sqrt(
        np.sum((np_vector_a - mean_a) ** 2)
    ) * np.sqrt(
        np.sum((np_vector_b - mean_b) ** 2)
    )

    return numerator / denominator


def cosine_similarity(
    np_vector_a: npt.NDArray,
    np_vector_b: npt.NDArray
) -> np.float64:
    # Compute dot product and magnitudes
    dot_prod = np.dot(np_vector_a, np_vector_b)
    magnitude_a = np.linalg.norm(np_vector_a)
    magnitude_b = np.linalg.norm(np_vector_b)

    return dot_prod / (magnitude_a * magnitude_b)


def test_pearson(
    np_vector_a: npt.NDArray,
    np_vector_b: npt.NDArray
) -> None:
    custom_corr = pearson_correlation(np_vector_a, np_vector_b)
    numpy_corr = np.corrcoef(np_vector_a, np_vector_b)[0, 1]

    if np.isclose(custom_corr, numpy_corr):
        print("Pearson correlation test passed")
    else:
        print("Pearson correlation test failed")


def test_cosine(
    np_vector_a: npt.NDArray,
    np_vector_b: npt.NDArray
) -> None:
    custom_cosine = cosine_similarity(np_vector_a, np_vector_b)
    scipy_cosine = 1 - spatial.distance.cosine(np_vector_a, np_vector_b)

    if np.isclose(custom_cosine, scipy_cosine):
        print("Cosine similarity test passed")
    else:
        print("Cosine similarity test failed")


# Test vectors
mean_centered_vectors = (
    np.array([-1.0, 0.0, 1.0]),
    np.array([-2.0, 0.0, 2.0])
)

non_mean_centered_vectors = (
    np.array([1.0, 2.0, 3.0]),
    np.array([2.0, 4.0, 6.0])
)

test_cases = [
    mean_centered_vectors,
    non_mean_centered_vectors
]

# Run tests
for vec_a, vec_b in test_cases:
    test_cosine(vec_a, vec_b)
    test_pearson(vec_a, vec_b)
````
